# Merge and data frame merging

 combines DataFrames based on a common key or index, similar to SQL joins.

```python
var1 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "B": [1, 2, 3, 4, 5, 6]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [12, 23, 45, 34, 34, 43]})

pd.merge(var1, var2, on="A")

```

```python
var1 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "B": [1, 2, 3, 4, 5, 6]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [12, 23, 45, 34, 34, 43]})

pd.merge(var1, var2, how="inner")

```

```python
var1 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "B": [1, 2, 3, 4, 5, 6]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [12, 23, 45, 34, 34, 43]})

pd.merge(var1, var2, left_index=True, right_index=True)
```

```python


var1 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "B": [1, 2, 3, 4, 5, 6]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [12, 23, 45, 34, 34, 43]})

pd.merge(var1, var2, left_index=True, right_index=True, suffixes=('name', 'mahi'))

```

# Concat

stacks DataFrames vertically or horizontally without considering relationships, while

```python
var1=pd.Series([12,23,4,4,5,2])
var2=pd.Series([43,54,43,3,12,4])
pd.concat([var1,var2])
```

```python
var1 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "B": [1, 2, 3, 4, 5, 6]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [12, 23, 45, 34, 34, 43]})

pd.concat([var1, var2])
```

```python
var1 = pd.DataFrame({"A": ["a", "b", "e", "f"], "B": [1, 2,4, 5]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [22,23, 45, 34, 34, 43]})

pd.concat([var1, var2],axis=1)
```

```python
var1 = pd.DataFrame({"A": ["a", "b", "e", "f"], "B": [1, 2,4, 5]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [22,23, 45, 34, 34, 43]})

pd.concat([var1, var2],join ='inner')
```

```python
var1 = pd.DataFrame({"A": ["a", "b", "e", "f"], "B": [1, 2,4, 5]})
var2 = pd.DataFrame({"A": ["a", "b", "c", "d", "e", "f"], "D": [22,23, 45, 34, 34, 43]})

pd.concat([var1, var2],join='outer')
```

```python
var0 = pd.DataFrame({"A": ["a", "b", "e", "f"], "B": [1, 2,4, 5]})
var3 = pd.DataFrame({"A": ["a", "b", "c", "d"], "D": [22,23, 45, 34]})

pd.concat([var0, var3],axis=1,keys=["var0","var3"])
```

```python
var0 = pd.DataFrame({"A": ["a", "b", "e", "f"], "B": [1, 2,4, 5]})
var3 = pd.DataFrame({"A": ["a", "b", "c", "d"], "D": [22,23, 45, 34]})

pd.concat([var0, var3],axis=0,keys=["var0","var3"])
```

# GroupBy

n Pandas, the groupby() function is used to split a DataFrame into groups based on the values in one or more columns. This allows you to perform operations on each group independently,

```python


df = pd.DataFrame({'City': ['New York', 'New York', 'Los Angeles', 'Chicago'],
        'Sales': [100, 200, 150, 300]})
grouped = df.groupby('City')
```

```python
for x,y  in grouped:
    print(x)
    print(y)
```

```python
grouped.min()
```

```python
grouped.max()
```

# Pivot Table

A pivot table is a data analysis tool that allows you to take columns of raw data from a pandas DataFrame, summarize them, and then analyze the summary data to reveal its insights.

```python

data = pd.DataFrame({'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
        'Region': ['North', 'South', 'North', 'South'],
        'Sales': [100, 150, 200, 120]})

pivot_table = pd.pivot_table(data, index='Date', columns='Region', values='Sales', aggfunc='sum')

print(pivot_table)
```

# Melt

Pandas melt() and unmelt using pivot() function. Pandas melt() function is used to change the DataFrame format from wide to long. 

```python


df = pd.DataFrame({'id': [1, 2], 'A': [3, 4], 'B': [5, 6]})

melted_df = df.melt(id_vars='id', value_vars=['A', 'B'], var_name='column', value_name='value')

print(melted_df)
```
